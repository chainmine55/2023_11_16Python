```python
import pandas as pd

scores = [
    {'chinese':78,'english':80,'math':92,'history':75,'discover':85},
    {'chinese':92,'english':60,'math':85,'history':62,'discover':73}
]
dataFrame = pd.DataFrame(data=scores,index=['徐國堂','徐家家'])
display(dataFrame.T)
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

#codespace
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

figure = plt.figure()
axes = figure.add_subplot(1,1,1)
dataFrame.T.plot(ax=axes,style=['--','-'],marker='o')
axes.set_title('徐國堂和徐家家分數表',fontproperties=font)
axes.grid(visible=True,axis='both')
plt.show()
```
